Загрузим датасет о ценах на дома в Бостоне. 

```python
from sklearn.datasets import load_boston
boston = load_boston()
print(boston.data.shape)
```

Каждый дом - это вектор в 13-мерном пространстве. 
Написать код для подсчета скалярного произведения двух первых домов из датасета.

```python
x1 = boston.data[0] # дом 1
x2 = boston.data[1] # дом 2
```

```python
dot_x1x2 = 0
for a,b in zip(x1,x2):
    dot_x1x2 += a*b
print('Сколярное произведение двух первых домов равно {}'.format(dot_x1x2))
```

этап 2: взять два атрибута и посчитать скалярное произведение двух векторов (2 дома), начертить 2 дома как вектора в двухмерном пространстве, и сравниваем графически и через скалярное произведение.


```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
dot2 = x1[0]*x2[0] + x1[2]*x2[2]
print('Скалярное произведение векторов по первому и третьему признакам равно {}'.format(dot2))
plt.plot([0, x1[0]],[0,x1[2]])
plt.plot([0, x2[0]],[0,x2[2]])
```

Найти косинус угла между векторами x1 и x2, где x1 и x2 - это два дома из 2ух атрибутов, полученных на этапе 2.

```python
import math
cosa = math.acos(dot2/(math.sqrt(x1[0]**2+x1[2]**2)*math.sqrt(x2[0]**2+x2[2]**2)))
print('Косинусное расстояние между векторами равно {}'.format(cosa))
```

Найти расстояние между точками x1 и x2.

```python
import math
dist = math.sqrt((x1[0]-x2[0])**2 + (x1[2]-x2[2])**2)
print('Расстояние между точками х1 и х2 равно {}'.format(dist))
```

# Задание повышенного уровня.

Задан объект в пространстве 2D. 
Изменить положение объекта (сместить по оси x на 300 пикселей, по оси y на 200 пикселей).

```python
import numpy as np
import matplotlib.pyplot as plt

A = [
    [-100, -100, 1],
    [-100, 100, 1],
    [100, 100, 1],
    [0, 0, 1],
    [100, -100, 1],
    [-100, -100, 1]
    ]
A = np.array(A)
x = A[:,0]
y = A[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

```python
TransformArray = np.array([
    [1, 0, 300],
    [0, 1, 200],
    [0, 0, 1]
])
```

```python
B = []
for point in A:
    new_point=np.dot(TransformArray,np.transpose(point))
    B.append(new_point)
B = np.array(B).astype(int)
```

```python
x = B[:,0]
y = B[:,1]
plt.plot(x, y)
plt.ylim([-200, 500])
plt.xlim([-200, 400])
```
